```python
import pandas as pd
import numpy as np
%matplotlib inline
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv("Park_file.csv")
#pd.set_option('display.max_rows', None)
data.dropna()
data[['name','neighborhood']].head(10)
```

```python
data["neighborhood"].value_counts().sort_index().plot(kind='bar', rot=0)
plt.title("Park Data")
plt.xlabel("Neighborhood")
plt.ylabel("Parks Per Neighbordhood")
plt.show()
```

```python
# top 10
data["neighborhood"].value_counts().sort_values(ascending = False).head(10).plot(kind='bar', rot=0)
plt.title("Top 10 Neighborhoods With Most Parks")
plt.xlabel("Neighborhood")
plt.ylabel("Parks Per Neighbordhood")
plt.show()
```
